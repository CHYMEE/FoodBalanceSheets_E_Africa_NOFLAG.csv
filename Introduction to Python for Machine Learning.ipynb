### QUESTION 1: What is the total Protein supply quantity in Madagascar in 2015?

```python
#filtering the rows for protein supplies in madagascar
protein_supply_madagascar_2015 = df_food[(df_food['Area'] == 'Madagascar') & (df_food['Element'] == "Protein supply quantity (g/capita/day)")]
protein_supply_madagascar_2015
```

```python
madagascar_2015 = 'Y2015'
```

```python
 #calculating the sum total of protein supplies made in the year 2015 in madagascar
total_protein_supply_madagascar_2015 = protein_supply_madagascar_2015[madagascar_2015].sum(). round(2)
total_protein_supply_madagascar_2015
```

### Which year had the least correlation with ‘Element Code’?

```python
# Calculate the correlation between 'Element Code' and each year column
correlations = df_food.iloc[:, 4:].corr().loc['Element Code']
correlations
```

```python
# Find the year with the least correlation
year_with_least_correlation = correlations.idxmin()
year_with_least_correlation
```

### Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
# Select columns 'Y2017' and 'Area'
selected_columns = ['Y2017', 'Area']
df_selected = df_food[selected_columns]
```

```python
# Perform a groupby operation on 'Area' and calculate the sum for 'Y2017'
area_grouped = df_selected.groupby('Area')['Y2017'].sum()
area_grouped
```

```python
# Find the Area with the 7th lowest sum in 2017
seventh_lowest_area = area_grouped.nsmallest(7).index[-1]
seventh_lowest_area 
```

### Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

```python
#Perform a groupby operation on ‘Element’
element_grouped = df_food.groupby('Element').sum()
element_grouped
```

```python
# Filter the DataFrame to rows where 'Element' is 'Stock Variation'
stock_variation_df = df_food[df_food['Element'] == 'Stock Variation']
stock_variation_df
```

```python
# Calculate the sum for each year
yearly_sum = stock_variation_df.drop(columns=['Element', 'Area', 'Area Code', 'Item Code', 'Item','Unit','Element Code']).sum()
yearly_sum
```

```python
# Find the year with the highest sum of Stock Variation
year_with_highest_stock_variation = yearly_sum.idxmax()
year_with_highest_stock_variation
```
